```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.random.random((20,100))
```

```python
plt.imshow(x)
```

```python
x = np.linspace(0, 4*np.pi, 100)
ys = np.array([np.sin(x)+np.random.randn(100) for _ in range(20)])
```

```python
plt.imshow(ys, 
           aspect='auto', 
           interpolation='nearest', 
           vmin=-4, 
           vmax=4, 
           cmap='PiYG_r')
plt.colorbar()
```

Colorbar smaller!

```python
plt.figure()
plt.imshow(ys, 
           aspect='auto', 
           interpolation='nearest', 
           vmin=-4, 
           vmax=4, 
           cmap='PiYG_r')

cb = plt.colorbar(shrink=.3)
cb.set_ticks([-4, 0, 4])
```
