```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
# %matplotlib notebook
```

Generate some fake data

```python
x = np.linspace(0, 100 * np.pi, 1000)
y = np.sin(x)
```

```python
# Magic to make real PDF text
plt.rcParams["svg.fonttype"] = 'none'
```

```python
plt.plot(x, y, color="k")
plt.savefig("plot_image.svg", dpi=300)
plt.savefig("plot_image.pdf", dpi=300)
plt.savefig("plot_image.eps", dpi=300)
```
