```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
# %matplotlib notebook
```

Generate some fake data

```python
x = np.linspace(0, 10 * np.pi, 100)
y = np.sin(x)
```

```python
# Magic to make real PDF text
plt.rcParams["svg.fonttype"] = 'none'
plt.rcParams["xtick.labelsize"] = 8
plt.rcParams["ytick.labelsize"] = 8
plt.rcParams["axes.labelsize"] = 8
plt.rcParams["legend.fontsize"] = 8
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(2., 1.5))
plt.plot(x, y, color="k", marker=".", label="data", alpha=.5, markersize=2)
plt.plot(x+1, y, color="k", marker=".", label="data", alpha=.5, markersize=2)
plt.xlabel("Time")
plt.ylabel("Amplitude [V]")
plt.legend(loc=[1.1, 0.1], frameon=False, shadow=None)

#          LOCATION             TEXT
plt.yticks([-1, -.5, 0, .5, 1], [-10, -5, 0, 5, 10])
plt.yticks([-1, -.5, 0, .5, 1], ["A", "B", "C", "D", "E"])

plt.xticks(np.arange(0, 32+1, 8))

plt.tight_layout()

sns.despine(trim=True)

plt.savefig("plot_image.svg", dpi=300)
plt.savefig("plot_image.pdf", dpi=300)
plt.savefig("plot_image.eps", dpi=300)
```

```python
x = np.linspace(0, 4*np.pi, 100)
ys = [np.sin(x)+np.random.randn(100) for _ in range(20)] # list comprehension

for y in ys:
    plt.plot(x,y)
```

```python
ys = np.array(ys)
```

```python
type(ys), ys.dtype, ys.shape
```

```python
m = ys.mean(0)
m.shape
```

```python
# Plot single data
plt.plot(np.repeat(x[None], 20, 0).T, ys.T, color='k', alpha=.06);
# Plot the mean
plt.plot(x, m, color="k")
```

```python
# Plot the mean
plt.plot(x, m, color="k")

std = ys.std(0)

plt.fill_between(x, m-std, m+std, alpha=.3, color='k')

plt.plot(x, m+np.linspace(0, 5, 100), color="m")
plt.fill_between(x, 
                 m+np.linspace(0, 5, 100)-std, 
                 m+np.linspace(0, 5, 100)+std, 
                 color='m', 
                 alpha=.3)
```
